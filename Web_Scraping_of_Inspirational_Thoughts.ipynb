```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
r= requests.get('https://www.brainyquote.com/topics/inspirational-quotes')
soup= BeautifulSoup(r.content,'html5lib')
```

```python
divh =soup.find('div',attrs={'id':'qbc1'})
print(divh)
```

```python
quotes_box=divh.findAll('div',attrs={'class':'grid-item qb clearfix bqQt b-qt-lg'})
quotes_box
```

```python
thoughts=[]
for quote in quotes_box:
    thoughts.append(quote.find('a').text)
```

```python
import pandas as pd
import numpy as np
```

```python
df =pd.DataFrame(data=thoughts,columns=['Inspirational thoughts'])
```

```python
df.replace('\n\n',np.NAN,inplace=True)
df.dropna(inplace=True)
df
```

```python
file = df.to_csv('inspirational_thoughts.csv',index=False)
```
